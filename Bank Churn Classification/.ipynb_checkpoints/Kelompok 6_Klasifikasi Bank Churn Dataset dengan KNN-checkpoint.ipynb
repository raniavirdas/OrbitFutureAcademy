#### Anova Analysis (Data Numerik)

- analisis variance anova - F-test
- digunakan untuk membandingkan mean (rata-rata) dari >= 2 sampel. Untuk memeriksa apakah setiap mean itu mirip atau tidak
- contoh nya: CreditScore = Anova untuk membandingkkan mean dari credit score yang meninggalkan bank dan yang bertahan

#### Uji hipotesis
- Ho : mean dari kedua grup itu mirip
- Ha : mean dari kedua grup berbeda satu sama lain
    
Untuk CI (condfidential interval), 95% sehingga Alpha 100-95 = 5% = 0.05
- Jika p-value < alpha = Terima Ha/ Tolak H0, yang artinya feature dianggap penting
- Jika p-value > alpha = Terima Ho/ Tolak Ha, yang artinya kedua feature itu tidak penting

#### Memeriksa P-value pada setiap feature/kolom terhadap kolom target class (Exited)
kita bisa menggunakan metode pada label numerik dan kategorik

```python
from sklearn.feature_selection import f_classif
```

```python
numeric_col = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
```

```python
xd = df[numeric_col]
yd = df['Exited']
```

```python
fvalue, pvalue = f_classif(xd, yd)
```

```python
for i in range(len(numeric_col)):
    print(numeric_col[i], pvalue[i], fvalue[i])
```

Berdasarkan analisa uji F diatas, dapat disimpulkan bahwa CreditScore, Age, Balance, dan NumofProducts memiliki p  value < alpha yang artinya menolak H0. sehingga dapat dikatakan bahwa features tersebut dianggap penting. Sementara itu, Tenure dan EstimatedSalary memiliki p value > alpha yang artinya menolak Ha. Sehingga dapat dikatakan bahwa features tersebut tidak penting. 

#### Chi Square Test (Data Kategorik)

digunakan untuk membandingkan distribusi dari feature kategorik >= 2 grup

Uji Hipotesis
- Ho : distribusi kategorik dalam 2 grup itu mirip
- Ha : distribusi kategorik dalam 2 grup itu berbeda

Untuk CI (condfidential interval), 95% sehingga Alpha 100-95 = 5% = 0.05
- Jika p-value < alpha = Terima Ha/ Tolak H0, yang artinya feature dianggap penting
- Jika p-value > alpha = Terima Ho/ Tolak Ha, yang artinya kedua feature itu tidak penting

## 4. Spliting data

```python
list(df)
```

```python
# Kolom kategorikal adalah

# Geography
# Gender
# HasCrCard
# IsActiveMember
# Exited
```

```python
X = df.drop(columns='Exited')
y = df.Exited

X = pd.get_dummies(X, columns=['Geography', 'Gender', 'HasCrCard', 'IsActiveMember'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
df.head()
```

## 5. Implementasi model KNN Classifier

### Tanpa Tuning

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
```

```python
knn.score(X_train,y_train), knn.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=3)
cross_val_score(knn, X, y, cv=3).mean()
```

### Improvement melalui features scalling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
```

```python
knn.score(X_train_scaled, y_train), knn.score(X_test_scaled, y_test)
```

#### Tuning jumlah tetangga tanpa scaling

```python
neighbor = range(1,16,2)
train_score = []
test_score = []


for k in neighbor:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    train_score.append(knn.score(X_train, y_train))
    test_score.append(knn.score(X_test, y_test))
    
plt.plot(neighbor, train_score, label='train')
plt.plot(neighbor, test_score, label='test')
plt.legend()

print(f"Max test score: {np.max(test_score)}")
print(f"n neighbor terbaik: {neighbor[np.argmax(test_score)]}")
```

#### Tuning Jumlah Tetangga dengan scaling

```python
neighbor = range(1,16,2)
train_score = []
test_score = []


for k in neighbor:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score.append(knn.score(X_train_scaled, y_train))
    test_score.append(knn.score(X_test_scaled, y_test))
    
plt.plot(neighbor, train_score, label='train')
plt.plot(neighbor, test_score, label='test')
plt.legend()

print(f"Max test score: {np.max(test_score)}")
print(f"n neighbor terbaik: {neighbor[np.argmax(test_score)]}")
```

###  Tuning jumlah tetangga dengan weight distance

#### Dengan euclidean distance

```python
neighbor = range(1,16,2)
train_score = []
test_score = []


for k in neighbor:
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance')
    knn.fit(X_train_scaled, y_train)
    
    train_score.append(knn.score(X_train_scaled, y_train))
    test_score.append(knn.score(X_test_scaled, y_test))
    
plt.plot(neighbor, train_score, label='train')
plt.plot(neighbor, test_score, label='test')
plt.legend()

print(f"Max test score: {np.max(test_score)}")
print(f"n neighbor terbaik: {neighbor[np.argmax(test_score)]}")
```

#### dengan manhattan distance

```python
neighbor = range(1,16,2)
train_score = []
test_score = []


for k in neighbor:
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance', p =1)
    knn.fit(X_train_scaled, y_train)
    
    train_score.append(knn.score(X_train_scaled, y_train))
    test_score.append(knn.score(X_test_scaled, y_test))
    
plt.plot(neighbor, train_score, label='train')
plt.plot(neighbor, test_score, label='test')
plt.legend()

print(f"Max test score: {np.max(test_score)}")
print(f"n neighbor terbaik: {neighbor[np.argmax(test_score)]}")
```

Dari hasil diatas dapat disimpulkan bahwa dengan menerapkan jumlah tetangga terdekat (k) = 11, scaling dengan min max scaler, dan weighting berupa euclidean distance diperoleh nilai akurasi  tertinggi pada data uji yaitu 84,45%.
